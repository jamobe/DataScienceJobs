```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
import os.path
path = os.getcwd()
parent_folder, current_folder = os.path.split(path)
parent_folder
```

```python
df = pd.read_csv(parent_folder+'/data/indeed_us_2019_11_05_10_02.csv', sep='\t', index_col=0)
```

```python
print(df.shape)
```

```python
df
```

```python
df['extraction_date'] = datetime(2019, 11, 5)
```

```python
df = df.rename(columns={"title": "job_title"})
df.company = df.company.str.strip()
df.description = df.description.str.strip()
```

```python
df = df.drop(['description'], axis=1)
```

```python
df.salary = df.salary.str.strip()
df['salary_low'] = df['salary'].str.split('-', n = 1, expand = True)[0]
df['salary_high'] = df['salary'].str.split('-', n = 1, expand = True)[1]
```

```python
df['job_type'] = 'Not available'
df['industry'] = 'Not available'
df['education'] = 'Not available'
df['career'] = 'Not available'
```

```python
df.to_csv(parent_folder + '/data/indeed_us_all.csv', index=True, sep='\t')
```
