```python
import pandas as pd
import plotly.graph_objects as go
import numpy as np
```

```python
df = pd.read_csv('UMAP.csv')
```

```python
df.columns
```

```python
df.color.value_counts()
```

```python
red = df.loc[df['label']=='Data Scientist']
blue = df.loc[df['label']=='Data Analyst']
green = df.loc[df['label']=='Data Engineer']
black = df.loc[df['label']=='Data Architect']
grey = df.loc[df['label']=='unknown']
```

```python
trace0 = dict(type='scatter', x=red.x, y=red.y, mode = 'markers', marker = dict(color='red'), text=red['label'], name = 'Data Scientist')
trace1 = dict(type='scatter', x=blue.x, y=blue.y, mode = 'markers', marker = dict(color='blue'), text=blue['label'], name = 'Data Analyst')
trace2 = dict(type='scatter', x=green.x, y=green.y, mode = 'markers', marker = dict(color='green'), text=green['label'], name = 'Data Engineer')
trace3 = dict(type='scatter', x=black.x, y=black.y, mode = 'markers', marker = dict(color='black'), text=black['label'], name = 'Data Architect')
trace00 = dict(type='scatter', x=grey.x, y=grey.y, mode = 'markers', marker = dict(color='lightgrey'), text=grey['label'], name = 'Unknown')

# Layout
layout = dict(title='UMAP visualization for: job descriptions using TFIDF', xaxis=dict(title=''),yaxis=dict(title=''), width=900, height=900)
```

```python
fig = go.Figure()

fig = go.Figure(data=[trace00, trace0, trace1, trace2, trace3], layout=layout) 
fig.update_traces(textposition='top center')

fig.show()
```

```python
fig.write_html(file='umap.html', include_plotlyjs=False, full_html=False)
```

```python
fig.write_html(file='umap_alone.html', include_plotlyjs='cdn')
```
